```python
#Question 4

def cleandata(dataset):
    func = lambda x: str(x).strip() 
    dataset = dataset.applymap(func) #removes spaces and special characters in the values
    dataset = dataset.replace("?",np.nan) #removes question marks
    dataset = dataset.replace("nan",np.nan) #replaces string nan with numpy.nan
    return dataset
```

```python
cd = cleandata(dataset)
```

```python
cd.dm.unique()
```

```python
#Question 5
#converting data types of following columns from object to float
col = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
for i in col:
    cd[i] = cd[i].astype(float) 
```

```python
cd.dtypes
```

```python
#Question 6

norm = 0
abnorm = 0

for i in range(len(cd.rbc)):
    if(cd['class'][i]=="ckd"): #if class is ckd
        if(cd['rbc'][i]=="normal"): #and rbc is normal
            norm+=1                 #increment normal count
        if(cd['rbc'][i]=="abnormal"): #if rbc is abnormal
            abnorm+=1                #increment abnormal count
print ("Normal RBC: "+str(norm))
print ("Abormal RBC: "+str(abnorm))
```

```python
#plots count of normal and abnormal values where class is ckd

cd[cd["class"]=="ckd"].rbc.value_counts().plot(kind='bar') 
plt.show()
```

```python
#Question 7
#finding max of bp where class is ckd

max = 0
for i in range(len(cd.bp)):
    if(max<cd.bp[i]and cd['class'][i]=="ckd"):
        max = cd.bp[i]
max
```
